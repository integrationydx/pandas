```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
pd.concat([df1,df2]) # this takes the input using the list hence in syntax we had to use [] also by default it does the concat column wise
```

```python
pd.concat([df1,df2],axis=1) # this will concat using the rows 
```
