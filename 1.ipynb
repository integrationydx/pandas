```python
import pandas as pd
import numpy as np
```

A series is a 1D array which can have any kind of data types inside it any data that is stack in vertical order is called series 

when a series is stacked horizontally stacked its called array

```python
d = {1:10,2:20,3:30}
arr = np.array([10,20,30])
my_list = [1,2,3]
labels = ['a','b','c']
```

```python
pd.Series(my_list) # here 0,1,2 are the indices which are called labels in here 
```

we can also send custom indexes just give the parameter in there

```python
pd.Series(my_list,index=labels) # here we got our custom labels and we can create series using a dictionary and arrays as well
```

```python
pd.Series(arr)
```

```python
pd.Series(d)
```
