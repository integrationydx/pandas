```python
import numpy as np
import pandas as pd
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [2, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.isna() # wherever there is null there it shows false and true otherwise 
```

```python
df.isna().sum()
```

```python
df.isna().any() # this tells us if the all the columns are filled if all the columns are not filled then it returns false 
```

```python
df.dropna() # the dropna is usually used to drop the data which is null and it works row wise as in here only 0th row had no NULL values hence only it survived 
```

```python
df.dropna(thresh=4) # tbis means only drop those rows which have minimum of 4 Non NULL values in themv 
```

```python
df.fillna(12) # fills the value NULL values with 12 
```

```python
# to fill values we can also first define the set of values that we wanna set # 
values = {"A":40,"B":30}
df.fillna(value =values) # fillna works column wise , drop occurs row wise , selection occurs column wise 
```
