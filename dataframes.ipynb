```python
import numpy as np
import pandas as pd
```

Multiple series together are called as Dataframes 

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
```

```python
df = pd.DataFrame(data) # we can create dataframes just by passing data to the Dataframe attribute 
```

```python
df # df is usally used to print the complete dataframe
```

```python
df2 = pd.DataFrame(df)
df2
```

```python
# we can also set columns individually
columns = ["Name","Age","City","Salary"]
df3 = pd.DataFrame(data, columns=columns)
df3
```

```python
# selecting a column, similarly we can select a row as well 
df3[["Name"]]
```

```python
# we can create new column as well 
df3["Designation"] = ["Doc","engg","consultant","arch"]
df3 # like this we can add new columns as well 
```

```python
# deleting a column 
df3.drop('Designation', axis=1, inplace=True, errors='ignore')
# if we dont do the inplace the changes wont propagate to the actual dataframe hence we set inplace = True
```

```python
df3.loc[0]  # the loc is used to select the rows in the table we cant directly access them we must use loc 
```

```python
df3.loc[[0,1]][["City","Salary"]] # so in here we selected a subset first we selected the the rows 0,1 and then we selected the city and salary also
```

```python
df3.loc[[2,3]][["Name","Age"]]
```

Conditional selection

```python
# selecting those whose age > 30
df3[df3["Age"]>30] # this gives us only those whose age was > 30
```

```python
# selecting age > 30 and city paris
df3[(df3["Age"]>30) & (df3["City"]== 'Paris')]
```
