```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}

df = pd.DataFrame(data)
```

```python
df
```

```python
cat = df.groupby('Category') # this way we can select the entire Category column
```

```python
cat = df.groupby('Category')['Sales'].sum() # this way we selected the category and sales and we did the .sum too 
cat 
```

```python
cat = df.groupby(['Category','Store'])['Sales'].sum()
cat 
```
