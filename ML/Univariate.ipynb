```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df
```

```python
# Plot the number of passengers in each survival class.
# In the Titanic data, Survived is usually coded as 0 = died and 1 = survived.
ax = sns.countplot(data=df, x="Survived")
ax.set_xticks([0, 1], ["Died", "Survived"])
ax.set_xlabel("Outcome")
ax.set_ylabel("Count")
ax.set_title("Passenger Survival Count")  # sns.countplot tells us how many survived and how many died basically 
# sns.countplot basically tells how many times each category is coming
```

```python
df['Pclass'].value_counts().plot(kind='pie',autopct = '%.2f') # this way we can generate the pie charts over here 
```
