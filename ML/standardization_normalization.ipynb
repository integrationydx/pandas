```python
from sklearn.preprocessing import StandardScaler

# StandardScaler standardizes each column by removing the mean and scaling to unit variance.
# Syntax breakdown:
#   df.drop(columns=['ID']) -> remove the identifier column
#   StandardScaler() -> create the scaler object
#   fit_transform(X) -> learn mean/std from X and transform it in one step
#   pd.DataFrame(..., columns=X.columns) -> convert the result back to a DataFrame
X = df.drop(columns=['ID'])

scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X_scaled.head()
```

```python
# above we performed standardisation
```

```python
from sklearn.preprocessing import MinMaxScaler

# MinMaxScaler normalizes each column to the 0-1 range.
# Syntax breakdown:
#   df.drop(columns=['ID']) -> remove the identifier column
#   MinMaxScaler() -> create the normalizer object
#   fit_transform(X_norm) -> learn min/max values and scale the data
#   pd.DataFrame(..., columns=X_norm.columns) -> keep the original column names
X_norm = df.drop(columns=['ID'])

normalizer = MinMaxScaler()
X_normalized = pd.DataFrame(normalizer.fit_transform(X_norm), columns=X_norm.columns)
X_normalized.head()
```
