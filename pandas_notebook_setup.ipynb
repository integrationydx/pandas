## 7. Import libraries and basic pandas settings

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.options.display.max_rows = 100
pd.options.display.max_columns = 50
```

## 8. Load sample datasets from CSV, Excel, or JSON

```python
df = pd.read_csv('data/sample.csv')
df_excel = pd.read_excel('data/sample.xlsx', sheet_name=0)
df_json = pd.read_json('data/sample.json')
df_dates = pd.read_csv('data/sample.csv', parse_dates=['date_col'])
```

## 9. Inspect and explore a DataFrame

```python
df.head()
df.tail()
df.info()
df.describe(include='all')
df.dtypes
df.sample(5)
```

## 10. Common transformations

```python
df_filtered = df[df['col'] > 0]
agg = df.groupby('key').agg({'val': 'sum'})
merged = df.merge(df[['id']].drop_duplicates(), on='id', how='left')
pivot = df.pivot_table(index='idx', columns='col', values='val', aggfunc='mean')
```

## 11. Simple plotting with pandas, matplotlib, and seaborn

```python
%matplotlib inline
df['col'].hist()
plt.show()
df.plot(kind='line', x='date', y='value')
plt.show()
sns.barplot(data=agg.reset_index(), x='key', y='val')
plt.show()
```
