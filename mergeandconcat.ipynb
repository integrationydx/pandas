```python
import numpy as np
import pandas as pd
```

```python

# DataFrame 1: Employee information
df1= pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
df2 = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
pd.merge(df1,df2,how = 'inner') # this merges the both the data sets the how = inner merges based on the common rows 
```

```python
pd.merge(df1,df2,how = 'outer')  # in salary 3 and 4 were not there hence it puts nan in there the method outer 
```

```python
pd.merge(df1,df2,how = 'left')  # we are merging based on df1 as its on the left
```

```python
pd.merge(df1,df2,how = 'right')  # we are now merging based on df2
```
